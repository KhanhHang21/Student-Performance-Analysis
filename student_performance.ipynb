## Anova test
**first categorize the Parental Support into five groups:**

1. None

2. Low

3. Moderate

4. High

5. Very High

**Compare the mean GPA across these five groups**  to see if there are statistically significant differences

**Steps:**
1. Create Parental Support Groups: We will categorize Parental Support into the five groups mentioned.

2. Group the Data: We will group GPA by these five support categories.

3. Perform ANOVA: We will use ANOVA to determine if there are significant differences in GPA across these five levels of Parental Support.


```python
from scipy.stats import f_oneway

# Create Parental Support Groups
df['ParentalSupportGroup'] = pd.cut(df['ParentalSupport'],
                                    bins=[-1, 1, 2, 3, 4, 5],
                                    labels=['None', 'Low', 'Moderate', 'High', 'Very High'])

# Group GPA by Parental Support Groups
grouped_gpa = [df[df['ParentalSupportGroup'] == group]['GPA'] for group in df['ParentalSupportGroup'].unique()]

# Perform ANOVA test
f_stat, p_value = f_oneway(*grouped_gpa)

# Display the results
f_stat, p_value
```

**Interpretation:**

 F-statistic: The F-statistic of **27.38 is quite large, indicating that there is a significant difference in GPA across the different levels of Parental Support.**

 P-value: The P-value is **extremely small (less than 0.05), meaning that we can reject the null hypothesis. This indicates that there is a statistically significant difference in the mean GPA across the different levels of Parental Support.**

Conclusion:
Since the P-value is less than 0.05, we can confidently conclude that **Parental Support significantly affects GPA in this dataset. This means that students with higher levels of Parental Support tend to have higher GPA scores.**

## ANOVA Test

```python
from scipy.stats import f_oneway

# Subset df
df_eca = df[['Extracurricular', 'GPA']]

# Drop missing values
df_eca = df_eca.dropna()
# Group GPA values based on Extracurricular participation
group_yes = df_eca[df_eca['Extracurricular'] == 'Yes']['GPA'].dropna()
group_no = df_eca[df_eca['Extracurricular'] == 'No']['GPA'].dropna()

# Perform one-way ANOVA
stat, p_value = f_oneway(group_yes, group_no)

# Print the result
print(f"ANOVA F-statistic: {stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Subset df
df_eca = df[['Sports', 'GPA']]

# Drop missing values
df_eca = df_eca.dropna()
# Group GPA values based on Extracurricular participation
group_yes = df_eca[df_eca['Sports'] == 'Yes']['GPA'].dropna()
group_no = df_eca[df_eca['Sports'] == 'No']['GPA'].dropna()

# Perform one-way ANOVA
stat, p_value = f_oneway(group_yes, group_no)

# Print the result
print(f"ANOVA F-statistic: {stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

**Interpretation:**

 The P-value is less than 0.05, which indicates that we reject the null hypothesis. This means that **there is significant difference between GPAs for students participating in extracurricular activities (sports, music, volunteering) in this dataset.**

Based on this ANOVA test, there is a statistically significant relationship between participation in extracurricular participation and sports with GPA. This result suggests that **extracurricular activities do have a clear impact on academic performance** in this dataset.

## g) Study Time and GPA

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from tabulate import tabulate

def linear_regression(df, target, dependent_var, hue=None):
    """ Create a scatter plot with linear regression line and optional hue for color grouping."""

    # Set figsize
    plt.figure(figsize=(10, 7))

    # Scatter plot
    sns.scatterplot(
        data=df,
        x=dependent_var,
        y=target,
        hue=hue,
        palette='Set1',
        s=75,
        edgecolor='black',
        alpha=0.3
    )

    # Prepare data for regression
    X = df[dependent_var].values.reshape(-1, 1)
    y = df[target].values

    # Perform linear regression
    reg = LinearRegression()
    reg.fit(X, y)
    y_pred = reg.predict(X)

    # Plot regression line
    plt.plot(df[dependent_var], y_pred, color='red', linewidth=2,
             label=f'Best Fit Line: y = {reg.coef_[0]:.2f}x + {reg.intercept_:.2f}')

    # Metrics
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    # Titles and labels
    plt.title(f'{target} vs {dependent_var}', fontsize=14)
    plt.xlabel(dependent_var, fontsize=12)
    plt.ylabel(target, fontsize=12)

    if hue:
        plt.legend(title=hue, loc='upper left', frameon=True, borderpad=1, fancybox=True)
    else:
        plt.legend(loc='upper left', frameon=True, borderpad=1, fancybox=True)

    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # OLS regression using statsmodels
    X_sm = sm.add_constant(X)
    model = sm.OLS(y, X_sm)
    results = model.fit()

    # Extract regression results
    slope = reg.coef_[0]
    intercept = reg.intercept_
    p_values = results.pvalues
    conf_int = results.conf_int()

    # Display summary in table
    headers = ['Metric', 'Value']
    table = [
        ['Slope', f'{slope:.4f}'],
        ['Intercept', f'{intercept:.4f}'],
        ['Root Mean Squared Error', f'{rmse:.4f}'],
        ['R-squared', f'{r2:.4f}'],
        ['P-value (Intercept)', f'{p_values[0]:.4f}'],
        ['P-value (Slope)', f'{p_values[1]:.4f}'],
        ['95% CI (Intercept)', f'[{conf_int[0][0]:.4f}, {conf_int[0][1]:.4f}]'],
        ['95% CI (Slope)', f'[{conf_int[1][0]:.4f}, {conf_int[1][1]:.4f}]'],
    ]

    print(tabulate(table, headers, tablefmt="pretty"))
```

```python
linear_regression(df, 'GPA', 'StudyTimeWeekly')
```

## ANOVA Test:

**first categorize the Parental Education into five groups:**

1. None

2. High School

3. Some College

4. Bachelor's

5. Higher

**compare the mean GPA across these five groups** using the ANOVA test to determine if there are statistically significant differences.


**Steps:**
1. Create Parental Education Groups: We will categorize the Parental Education into the five groups mentioned.

2. Group the Data: We will group GPA by these five education categories.

3. Perform ANOVA: We will use ANOVA to see if there are significant differences in GPA across these five groups.

```python
from scipy.stats import f_oneway

# Create Parental Education Groups
df['ParentalEducationGroup'] = pd.cut(df['ParentalEducation'],
                                      bins=[-1, 0, 1, 2, 3, 4],
                                      labels=['None', 'High School', 'Some College', 'Bachelor\'s', 'Higher'])

# Group GPA by Parental Education Groups
grouped_gpa = [df[df['ParentalEducationGroup'] == group]['GPA'] for group in df['ParentalEducationGroup'].unique()]

# Perform ANOVA test
f_stat, p_value = f_oneway(*grouped_gpa)

# Display the results
f_stat, p_value
```

**Results**

**F-statistic:** 1.81

**P-value:** 0.1245

**Interpretation:**

F-statistic: The F-statistic of 1.81 suggests that **there is some variability in GPA across the different Parental Education groups, but the difference is not very large.**

P-value: The P-value of 0.1245 is greater than 0.05, meaning that **we fail to reject the null hypothesis. This indicates that there is no statistically significant difference in the mean GPA across the different levels of Parental Education.**

Conclusion:
Based on the ANOVA test, **Parental Education does not seem to have a significant effect on GPA** in this dataset. The difference in mean GPA across the five Parental Education groups is not statistically significant.

```python
from scipy import stats
from tabulate import tabulate

def chi_square_test_of_independence(df, categorical_variable_1, categorical_variable_2):
    """
    Perform a Chi-square test of independence/association between two categorical variables.

    Parameters:
    df (pd.DataFrame): DataFrame containing the categorical variables.
    categorical_variable_1 (str): The first categorical variable.
    categorical_variable_2 (str): The second categorical variable.

    Returns:
    None. Prints the Chi-square test results in a table format.
    """
    # Create a contingency table
    contingency_table = pd.crosstab(df[categorical_variable_1], df[categorical_variable_2])

    # Perform the Chi-square test
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

    # Create a metrics table with tabulate
    headers = ['Metric', 'Value']
    table = [
        ['Chi-Square Statistic', f'{chi2:.4f}'],
        ['p-value', f'{p:.4f}'],
        ['Degrees of Freedom', dof],
    ]

    # Print the results in a neat format
    print(tabulate(table, headers, tablefmt="pretty"))

    # Optional: Return the contingency table for further inspection
    return contingency_table
```
